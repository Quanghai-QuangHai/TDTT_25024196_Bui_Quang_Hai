Câu 2 (Ảnh 69)Đề bài: "Vừa gà vừa chó, bó lại cho tròn, ba mươi sáu con, một trăm chân chẵn".Từ bài toán dân gian trên, mở rộng thành chương trình nhận đầu vào là tổng số con ($n$) và tổng số chân ($m$) của gà và chó. Nếu tìm được số phù hợp, in ra số lượng gà và số lượng chó. Ngược lại, in ra "invalid".

```python
n = int(input())
m = int(input())
found = False
for ga in range(n + 1):
    cho = n - ga
    if cho * 4 + ga * 2 == m:
        print(cho)
        print(ga)
        found = True
        break
if not found:
    print("Invalid")
```

Đề bài: Cho hai số $m$ và $n$ là hai số nguyên dương nhập từ bàn phím. Viết chương trình tìm ước chung lớn nhất (ƯCLN) của hai số $m$ và $n$.

```python
m, n = map(int,input().split())
a = m
b = n
while a != b:
    if a > b:
        a = a - b
    else:
        b = b - a
print(a)
```

Bài 4: Cặp số thân thiết

```python
a, b = map(int,input().split())
tổng_a = sum(i for i in range(1,a) if a % i == 0)
tổng_b = sum(i for i in range(1,b) if b % i == 0)
if tổng_a == b and tổng_b == a:
    print("True")
else:
    print("False")
```

```python
Viết hàm kiểm tra cso phải là số nguyên tố không
```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
n = int(input())
print(is_prime(n))
```

Viết hàm đếm số ước chẵn của số nhỏ hơn 10*6

```python
n = int(input())
count = 0
for i in range(1, n +1):
    if n % i == 0 and i % 2 == 0:
        count += 1
print(count)
```

```python
def is_lucky(n):
    tong = sum(int(i) for i in str(n))
    if tong == 0: return False
    return n % tong == 0
n = int(input())
print(is_lucky(n))
```

tổng các số trong số nguyên dương n 

```python
def tong_cac_so(n):
    s = str(n)
    tong = 0
    for chu_so in s:
        tong += int(chu_so)
    return tong
n = int(input())
print(tong_cac_so(n))
```
