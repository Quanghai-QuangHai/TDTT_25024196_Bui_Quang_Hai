Bài 1: Nhập vào số nguyên n, in ra màn hình số gấp đôi của n 

```python
# Bước 1: Nhập dữ liệu từ bàn phím và ép kiểu sang số nguyên (int)
n = int(input("Nhập vào số nguyên n: "))

# Bước 2: Tính số gấp đôi
ket_qua = n * 2

# Bước 3: In kết quả ra màn hình
print(f"Số gấp đôi của {n} là: {ket_qua}")
```

Bài 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu
vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần
diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2
chữ số phần thập phân. 



```python
# Nhập dữ liệu từ bàn phím
a = float(input("Nhập chiều rộng a: "))
b = float(input("Nhập chiều dài b: "))

# Khai báo số Pi
pi = 3.14

# Tính diện tích hình chữ nhật
s_hcn = a * b

# Tính bán kính hình tròn (giả định đường kính bằng chiều rộng a)
r = a / 2

# Tính diện tích hình tròn
s_tron = pi * (r ** 2)

# Diện tích còn lại để trồng cây
s_con_lai = s_hcn - s_tron

# In kết quả, làm tròn 2 chữ số thập phân
print(f"Diện tích còn lại để trồng cây là: {s_con_lai:.2f}")
```

Bài 3: Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường
tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem
trước cấu trúc lựa chọn: if….else, ma) 


```python
# Nhập kí tự từ bàn phím
c = input("Nhập vào một kí tự: ")

# Kiểm tra xem kí tự có độ dài bằng 1 hay không (để tránh lỗi khi nhập cả chuỗi)
if len(c) == 1:
    if c.isupper():
        # Nếu là chữ hoa, chuyển thành chữ thường
        print(f"Kết quả: {c.lower()}")
    elif c.islower():
        # Nếu là chữ thường, chuyển thành chữ hoa
        print(f"Kết quả: {c.upper()}")
    else:
        # Trường hợp là số hoặc kí tự đặc biệt
        print("Đây không phải là chữ cái.")
else:
    print("Vui lòng chỉ nhập đúng một kí tự!")
```

Bài 4: Viết chương trình nhập vào một “kí tự”, Kiểm tra xem c có phải là kí tự chữ cái
không (hoa hoặc thường). Nếu đúng thì in ra “{c} là kí tu alphabet”, nếu sai in ra “{c}
không phải là kí tự alphabet. 


```python
# Nhập kí tự từ bàn phím
c = input("Nhập vào một kí tự: ")

# Kiểm tra độ dài để đảm bảo người dùng chỉ nhập 1 kí tự
if len(c) == 1:
    # Sử dụng phương thức isalpha() để kiểm tra chữ cái
    if c.isalpha():
        print(f"{c} là kí tự alphabet")
    else:
        print(f"{c} không phải là kí tự alphabet")
else:
    print("Vui lòng chỉ nhập duy nhất một kí tự!")
```

Bài 5: Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương
ứng của A. Xét TH khi c = ‘A’ là đặc biệt. 

```python
# Nhập kí tự hoa
C = input("Nhập vào một chữ cái hoa: ")

# Kiểm tra nếu đúng là chữ hoa
if 'A' <= C <= 'Z':
    # Chuyển về chữ thường tương ứng
    c_thuong = C.lower()
    
    # Xử lý trường hợp đặc biệt: 'A'
    if C == 'A':
        ket_qua = 'z'
    else:
        # Lấy kí tự liền trước bằng cách trừ đi 1 trong bảng mã ASCII
        ket_qua = chr(ord(c_thuong) - 1)
    
    print(f"Chữ cái thường liền trước của '{c_thuong}' là: '{ket_qua}'")
else:
    print("Vui lòng nhập một chữ cái viết HOA!")
```

Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 

```python
import math

# Nhập 3 cạnh từ bàn phím
a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))

# 1. Kiểm tra điều kiện tồn tại tam giác
if (a + b > c) and (a + c > b) and (b + c > a):
    # 2. Tính nửa chu vi p
    p = (a + b + c) / 2
    
    # 3. Tính diện tích theo công thức Heron
    s = math.sqrt(p * (p - a) * (p - b) * (p - c))
    
    # In kết quả lấy 1 chữ số thập phân
    print(f"{s:.1f}")
else:
    # Nếu không thỏa mãn điều kiện
    print("Khong phai 3 canh cua tam giac")
```

Bài 7: Nhập vào một chuỗi gồm các kí tự thường độ dài >=20, in ra chữ cái thứ năm và
thứ chín của chuỗi này. 


```python
# Nhập chuỗi từ bàn phím
s = input("Nhập vào một chuỗi (độ dài >= 20): ")

# Kiểm tra độ dài chuỗi theo yêu cầu đề bài
if len(s) >= 20:
    # Lấy kí tự tại vị trí thứ 5 (index 4) và thứ 9 (index 8)
    ki_tu_5 = s[4]
    ki_tu_9 = s[8]
    
    print(f"Chữ cái thứ năm là: {ki_tu_5}")
    print(f"Chữ cái thứ chín là: {ki_tu_9}")
else:
    print("Chuỗi quá ngắn, vui lòng nhập chuỗi có ít nhất 20 kí tự!")
```

Bài 8: Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này,
tính tiền điện tháng này cho hộ biết:
- Từ 0 – 50kWh: 1.984 đồng/kWh
- Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh
- Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
- Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
- Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh 


```python
# Nhập thông tin khách hàng
ho_ten = input("Ten chu ho: ")
chi_so_cu = int(input("Chi so thang truoc: "))
chi_so_moi = int(input("Chi so thang nay: "))

# Tính số điện tiêu thụ
so_dien = chi_so_moi - chi_so_cu

# Khởi tạo biến tiền điện (chưa thuế)
tien_dien = 0

# Tính toán theo bậc thang
if so_dien > 400:
    tien_dien += (so_dien - 400) * 3460
    so_dien = 400
if so_dien > 300:
    tien_dien += (so_dien - 300) * 3350
    so_dien = 300
if so_dien > 200:
    tien_dien += (so_dien - 200) * 2998
    so_dien = 200
if so_dien > 100:
    tien_dien += (so_dien - 100) * 2380
    so_dien = 100
if so_dien > 50:
    tien_dien += (so_dien - 50) * 2050
    so_dien = 50
if so_dien > 0:
    tien_dien += so_dien * 1984

# Tính thuế VAT 8%
tong_phai_tra = tien_dien * 1.08

# In kết quả theo đúng định dạng mẫu (làm tròn đến đơn vị đồng)
print(f"Ho va ten: {ho_ten}")
print(f"Tien phai tra la: {tong_phai_tra:.0f}")
```
